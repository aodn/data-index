```python
import polars
import pathlib
import rich
import json

jprint = lambda o: rich.print(json.loads(o))
```

```python
smlf = polars.scan_parquet(pathlib.Path("../.load/structured_metadata"))
umlf = polars.scan_parquet(pathlib.Path("../.load/unstructured_metadata"))
```

```python
display(smlf.head().collect())
display(umlf.head().collect())
```

```python
prefixes = umlf.select("prefix").unique().collect()["prefix"]
display(prefixes)
```

```python
for prefix in prefixes:
    print(prefix)
    df = smlf.filter(
        polars.col("prefix").eq(prefix),
    ).collect()
    display(df.describe())
```
